## Our first program: GC content

This opening example is biologically meaningful and compact.

It demonstrates:
- function definition
- variables
- a loop over a sequence
- a conditional statement
- a return value


```python
def GC_content(dna_string):
    ''' calculate %GC in dna_string (uppercase ACTG) '''
    GC = 0
    for nuc in dna_string: # iterate over dna_string
        if nuc == "C" or nuc == "G":
            GC += 1 # equal to GC = GC+1
    return 100*GC/len(dna_string)
```

```python
GC_content("GT")
```

```python
GC_content("CCG")
```

```python
GC_content("TTAGACCAGTAGAAGTAC")
```

## Main keypoints from the first example

### Structure of the function
- `def` starts a function definition
- indentation defines the function body
- the docstring describes the function

### Main logic
- scan the DNA string one nucleotide at a time
- count only `C` and `G`
- divide by total length
- multiply by 100 to get a percentage

### Good classroom questions
- What happens on an empty string?
- What happens on lowercase input?
- Could we write the same idea in a different way?


## Extra visualization: nucleotide histogram

A useful teaching extension is to connect the counting loop to a visual summary.

Run the cell below after explaining `GC_content`.


```python
sample_dna = "TTAGACCAGTAGAAGTAC"
counts = Counter(sample_dna)

bases = ["A", "C", "G", "T"]
values = [counts.get(base, 0) for base in bases]

plt.figure(figsize=(7, 4))
plt.bar(bases, values)
plt.title("Nucleotide counts in a sample DNA sequence")
plt.xlabel("Base")
plt.ylabel("Count")
plt.show()
```

### Teaching message

The original loop computes a single summary number: **%GC**.

This visualization shows the full distribution:
- how many `A`
- how many `C`
- how many `G`
- how many `T`

So students can see that a simple loop can support both:
- numerical results
- graphical summaries

## Extra visualization: slicing as a window on a sequence

This figure supports the idea of:
- start index
- end index
- extracted block


```python
organism = "drosophila"
start = 0
end = 4

plt.figure(figsize=(10, 2.6))
ax = plt.gca()
ax.set_xlim(-0.5, len(organism) - 0.5)
ax.set_ylim(0, 1)

for i, ch in enumerate(organism):
    y = 0.65 if start <= i < end else 0.35
    ax.text(i, y, ch, ha="center", va="center", fontsize=14)
    ax.text(i, 0.08, str(i), ha="center", va="center", fontsize=10)

ax.set_xticks([])
ax.set_yticks([])
ax.set_title(f'Slicing example: organism[{start}:{end}] -> "{organism[start:end]}"')
for spine in ax.spines.values():
    spine.set_visible(False)
plt.show()
```

## Conditional statements

Conditionals split the program into alternative execution paths.

A very clean example from the chapter:


```python
def state(Tm, Tv, T):
    ''' return the state of material with Tm and Tv at temperature T '''
    if T < Tm:
        return "solid"
    else:
        if T < Tv:
            return "liquid"
        else:
            return "gas"
```

```python
state(0, 100, 25)
```

### Main keypoints
- `if` checks a condition
- `else` gives the alternative branch
- indentation determines which statements belong to each branch
- this is a natural place to ask students to predict outputs before running the cell


## Iteration with `for` loops

The chapter introduces `for` loops as a natural way to iterate over strings and lists.


```python
def basic_aa_perc(prot):
    ''' return percentage of basic amino acids in protein '''
    basic = 0
    basic_aa_list = ["K", "H", "R"]
    for aa in prot:
        if aa in basic_aa_list:
            basic += 1
    return 100*basic/len(prot)
```

```python
basic_aa_perc("MHKRAGG")
```

## Iteration with `while` loops

A `while` loop repeats as long as a condition remains true.

This version rewrites GC content with an explicit running index.


```python
def GC_content3(dna_string):
    ''' calculate %GC in dna string (uppercase ACTG) '''
    GC = 0
    i = 0
    while i < len(dna_string):
        if dna_string[i] == "C" or dna_string [i] == "G":
            GC += 1
        i += 1
    return 100*GC/len(dna_string)
```

```python
GC_content3("TTAGACCAGTAGAAGTAC")
```

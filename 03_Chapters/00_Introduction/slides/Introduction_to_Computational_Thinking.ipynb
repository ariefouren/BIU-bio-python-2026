## **What Is This Course About?**
Modern biology is undergoing a dramatic revolution. Enormous amounts of data are produced at an unprecedented rate, including:
*   Gene and protein expression levels.
*   DNA, RNA, and protein sequencing.
*   High-quality biological and medical images.

**The Goal:** Bridging the cultural gap between biological and computational sciences by teaching **algorithmic and logical thinking** rather than just tool usage.

## **The Computational Thinking Pipeline**
Computational methods are not just for analysis; they are integrated into every stage of research:

1.  **Lab-Experiment Design:** Structuring research to be computationally viable.
2.  **Data Generation:** Capturing and processing high-throughput data.
3.  **Integration & Analysis:** Combining disparate datasets for biological insight.
4.  **Modeling & Simulation:** Conducting *in-silico* experiments to predict system behavior.

## **Learning Objectives**
The primary aim is to develop fundamental skills that are often missing from standard biology training:

*   **Acquaintance with "Computational Culture":** Moving beyond simple tool handling.
*   **Non-Formal Learning:** Topics are presented from a biological point of view in a user-friendly manner.
*   **Hands-on Practice:** Developing self-efficacy through trial-and-error coding.

## **Learning Outcomes: Skills Gained**
By the end of this course, students should be able to:

*   **Identify Amenable Problems:** Recognize manual tasks that are too tedious or error-prone for humans but perfect for code.
*   **Implement Solutions:** Build simple, functional Python scripts for biological challenges.
*   **Understand CS Notions:** Grasp how computer science concepts apply directly to life sciences.
*   **Collaborate Effectively:** Better communicate with professional programmers and computer scientists.

## **Introductory Code Snippet**
### Calculating GC Content
This simple example demonstrates how we can automate the analysis of a DNA sequence using basic loops and conditionals.

```python
def GC_content(dna_string):
    '''Calculate %GC in dna_string (uppercase ACTG)'''
    GC = 0
    for nuc in dna_string: # iterate over each nucleotide
        if nuc == "C" or nuc == "G":
            GC += 1
    return 100 * GC / len(dna_string)

# Example usage with a sequence from Mycobacterium tuberculosis
sequence = "TTAGACCAGTAGAAGTAC"
print(f"Sequence: {sequence}")
print(f"GC Content: {GC_content(sequence):.2f}%")
```

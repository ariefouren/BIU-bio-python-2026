## Chapter overview — main points

- Complexity is one of the most fundamental ideas in computer science.

  - memory usage

- The goal is intuition, not formal proof-heavy analysis.

## 2.1 Running Time of Algorithms — main points

- Code quality is not measured only by shortness.
- Shorter code does **not** necessarily mean a better algorithm.
- Efficiency usually concerns two resources:
  - time
  - memory
- One direct way to evaluate efficiency is to **measure actual running time**.
- Python provides `time.perf_counter()` for high-resolution timing.


## Original code snippet — measuring time

The following code snippet is kept carefully as it appears in the chapter.


````python
import time

def stopper():

    t0 = time.perf_counter()

    i = 0

    while i<10**6:

        i = i+1

    t1 = time.perf_counter()

    return t1-t0
````

## Why actual timing is not enough — main points

- Actual running time changes from one computer to another.
- Even on the same computer, timing can vary.
- Timing on one particular input tells little about other inputs.
- Therefore computer scientists study the **number of operations as a function of input size**.
- That leads to the idea of **time complexity**.


## 2.2 Linear and Quadratic Time Complexity — main points

- The important question is not only “how long did it run?”
- The deeper question is:
  **How does running time grow when input size grows?**
- A loop over all items in a list gives a **linear** pattern.
- A nested loop over all pairs gives a **quadratic** pattern.
- Growth rate matters more than one isolated timing measurement.


## Original code snippet — `linear`


````python
def linear(L):

    for i in range(len(L)):

        print(L [i])
````

## Linear growth — main points

- `linear(L)` prints every element once.
- If the list length doubles, the work roughly doubles.
- This is called **linear growth**.
- Graphically, the time-vs-input-size plot is approximately a straight line.

## Key interpretation — main points

- Expressions like `n`, `n/2`, `1000n`, and `17/2 n + 32` all grow linearly.
- They do **not** have the same actual running time.
- They do belong to the same **complexity class**.
- Linear time complexity is denoted **O(n)**.

## Original code snippet — `quadratic`


````python
def quadratic(L):

    for i in range(len(L)):

        for j in range(len(L)):

            print(L [i], L [j])
````

## Quadratic growth — main points

- `quadratic(L)` prints all ordered pairs.
- For length `n`, this gives about `n²` operations.
- If input size doubles, running time grows by about a factor of 4.
- This is called **quadratic growth**.

## Original code snippet — `quadratic_too`


````python
def quadratic_too(L):

    for i in range(len(L)):

        for j in range(i,len(L)): # note the change

            print(L [i], L [j])
````

## `quadratic_too` — main points

- This version removes about half the pairs.
- It is faster in practice than `quadratic`.
- But its dominant term is still quadratic.
- Therefore its asymptotic complexity remains **O(n²)**.

## Linear vs quadratic — comparison points

- For small inputs, constants can matter a lot.
- For large inputs, growth rate usually dominates.
- A linear algorithm will eventually beat a quadratic one for large enough input size.
- This is why asymptotic thinking is so important.


## 2.3 The O Notation — main points

- O notation is a compact way to describe complexity classes.
- `O(n)` means linear time.
- `O(n²)` means quadratic time.
- It focuses on the **dominant part** of the running-time expression.
- It is excellent for comparing growth rates on large inputs.


## O notation — limits and interpretation

- O notation hides constants and lower-order terms.
- This makes it powerful but coarse.
- Two algorithms in the same O-class can still differ greatly in actual speed.
- Good practice:
  - analyze asymptotic complexity
  - also measure actual running time


## 2.4 Best-Case, Worst-Case — main points

- Some algorithms depend not only on input size, but also on the input itself.
- Search is a classic example.
- Best case and worst case can differ significantly.
- Worst-case analysis is the most common in algorithm analysis.


## Original code snippet — `search`


````python
def search(L, x):

    for item in L:

        if item == x:

            return True

    return False
````

## `search` — interpretation

- Best case:
  - the element is first
  - only one comparison
  - **O(1)**
- Worst case:
  - the element is absent
  - all elements are checked
  - **O(n)**
- Average case is often harder to analyze.


## 2.5 Binary Search and Logarithmic Time Complexity — main points

- Binary search works on a **sorted** list.
- It repeatedly cuts the search interval roughly in half.
- This gives **logarithmic** time complexity.
- Logarithmic algorithms are far more efficient than linear or quadratic ones on large inputs.


## Original code snippet — `binary_search`


````python
def binary_search(L, x):

    left, right = 0, len(L)-1

    while left <= right:

        mid = (left+right)//2

        if L [mid] == x:

            return True

        elif L [mid]> x:

            right = mid-1

        else:

            left = mid+1

    return False
````

## `binary_search` — main points

- Start in the middle.
- If the middle is too large, search left half.
- If the middle is too small, search right half.
- Each iteration halves the remaining search region.
- Complexity: **O(log n)**.


## 2.6 Exponential Complexity and Tractability — main points

- Polynomial-time algorithms such as:
  - `O(log n)`
  - `O(n)`
  - `O(n²)`
  - `O(n³)`
  are considered **tractable**.
- Exponential-time algorithms such as `O(2ⁿ)` are usually **intractable**.
- Exponential growth becomes impractical very quickly.


## Preserved table — Table 2.2

**Table 2.2. Tractable and intractable complexity classes**

| Tractability | Complexity class | Maximal n for which running time <= 1 sec |
|---|---|---|
| Tractable | `log₂ n` | `n = 2^(10^10) = 2^10,000,000,000` |
| Tractable | `n` | `n = 10^10 = 10,000,000,000` |
| Tractable | `n²` | `n = sqrt(10^10) = 10^5 = 100,000` |
| Intractable | `2ⁿ` | `n = log₂(10^10) ≈ 33` |


## Exponential complexity — main points

- Even a very fast computer cannot handle large `n` when the complexity is exponential.
- Example from the chapter:
  printing all genomic sequences of length `n`.
- There are `4ⁿ` such sequences.
- Therefore the task is inherently exponential in output size.


## 2.7 Useful Tips — main points

- Built-in functions are useful, but not “free”.
- `min`, `sum`, reversing a list:
  often linear in input size.
- `sorted` is typically `O(n log n)`.
- Slicing a string or list copies data:
  slicing length `m` costs **O(m)**.

- Some algorithms depend on more than one input size.

## Biological context — main points

- In computational biology, inputs can be extremely large:
  - genomes
  - proteomes
  - large sequence collections
- In such settings, even quadratic algorithms can be too slow.
- This motivates searching for linear or near-linear methods whenever possible.


## Trust but check — main points

- An `O(n)` algorithm can still be slower than an `O(n²)` algorithm for small inputs.
- Constants hidden in O notation may dominate at small scales.
- Therefore:
  - theory gives growth behavior
  - timing experiments give practical behavior
- Both are useful.


## Reflection — summary points

- Complexity helps us predict whether an algorithm will scale.
- Time and memory are both important resources.
- Sometimes there is a tradeoff:
  - use more memory to save time
- An inefficient algorithm may be practically useless if it:
  - runs for too long
  - or causes memory overflow


## Challenge Yourself — repeating k-mer problem

Main idea:
- Given a string `st` and an integer `k`,
  determine whether a k-mer repeats.
- The chapter begins with a naïve solution
  and later points to a more efficient hashing-based solution.


## Original code snippet — `repeating_substring_naive1`


````python
def repeating_substring_naive1(st, k):

    n = len(st)

    for i in range(n-k+1):

        for j in range(i+1, n-k+1):

            if st [i:i+k] == st [j:j+k]:

                return st [i:i+k]

    return None
````

## Original code snippet — `repeating_substring_wrong`


````python
def repeating_substring_wrong(st, k):

    for i in range(len(st)-k+1):

        if st [i:i+k] in st:

            return st [i:i+k]

    return None
````
